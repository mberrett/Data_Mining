# Visualization for Q1

```python
import numpy as np
import matplotlib.pyplot as plt

N = 12

k_means = (41652, 29523.88, 24123.36, 20982.81, 18744.78, 17145.57, 15500.01, 14380, 12807, 12212.13, 11837.95, 10995.35)


k_std = (0, 2403.72, 577.59, 1365.57, 1766.7, 2102.4, 1846.04, 1448, 1514.1, 1338.84, 1194.24, 1097.34)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, k_means, width, color='w', alpha = 0.5, yerr=k_std)
ax.plot(ind, k_means, 'o-', color = 'r',alpha = 0.5)
# add some text for labels, title and axes ticks
ax.set_ylabel('Mean SSE Score')
ax.set_xlabel('K value')
ax.set_title('K-means cluster performance by k value')
ax.set_xticks(ind)
ax.set_xticklabels([str(x+1) for x in range(12)])
ax.set_ylim([0,50000])

plt.show()
```
