```r
require("plotrix") 
x <- c(0,1,1,2,3,5,4,5,6,7)
y <- c(0,2,6,3,4,1,2,3,2,4)
plot(x,y, col="blue",xlim=c(-2,9), ylim=c(-2,8))
drawCircle <- function(x,y) {
    for (k in 1:10 ) {
        draw.circle(x[k],y[k],sqrt(2), col=)
        text(x[k]+0.1,y[k]+0.15,paste("(",x[k],",",y[k],")"), cex=0.7)
    }
}
drawCircle(x,y)
a=0.25
arrows(x[4]-a,y[4]-a,x[2]+a,y[2]+a, length=0.1, angle=20, col="dark green")
arrows(x[4]+a,y[4]+a,x[5]-a,y[5]-a, length=0.1, angle=20, col="dark green")
arrows(x[6]-a,y[6]+a,x[7]+a,y[7]-a, length=0.1, angle=20, code = 3, col="blue")
arrows(x[6]+a,y[6]+a,x[9]-a,y[9]-a, length=0.1, angle=20, code = 3, col="blue")
arrows(x[9]-a,y[9]+a,x[8]+a,y[8]-a, length=0.1, angle=20, code = 3, col="blue")
arrows(x[7]+a,y[7]+a,x[8]-a,y[8]-a, length=0.1, angle=20, code = 3, col="blue")
```
